```python
import tensorflow as tf
import numpy as np
```

```python
tf.enable_eager_execution()
```

```python
X = np.array([1,2,3], dtype='float32')
X
```

```python
y = np.array([2.5,5,3.5], dtype='float32')
y
```

```python
w = tf.Variable([2], dtype='float32')
b = tf.Variable([1], dtype='float32')
```

```python
w
```

```python
b
```

```python
hypothesis = w*X+b
hypothesis
```

```python
tf.square(hypothesis-y)
```

```python
tf.reduce_mean(tf.square(hypothesis-y))
```

```python
learning_rate = 0.1
```

```python
for i in range(1000):
    with tf.GradientTape() as tape:
        hypothesis = w*X+b
        print('hypothesis : ',hypothesis)
        cost = tf.reduce_mean(tf.square(hypothesis-y))
        print('cost : ',cost)
    w_grad,b_grad = tape.gradient(cost,[w,b])
    print('w_grad : ',w_grad,'b_grad : ',b_grad)
    w.assign_sub(w_grad*learning_rate)
    b.assign_sub(b_grad*learning_rate)
    print('w : ',w)
    print('b : ',b)
    print('--'*40)
```

```python
print(w.numpy()[0])
print(b.numpy()[0])
print(cost.numpy())
```
