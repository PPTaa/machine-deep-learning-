```python
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
import numpy as np
from sklearn import datasets
from sklearn.model_selection import train_test_split
```

```python
data = datasets.load_iris()
```

```python
X=data.data
y=data.target
y=y.reshape(-1,1)
```

```python
X=np.array(X,dtype='float32')
y=np.array(y,dtype='float32')
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
model = Sequential()
```

```python
X_train.shape
```

```python
model.add(Dense(16,input_dim=4, activation="sigmoid"))
```

```python
model.add(Dense(3,activation='softmax'))
```

```python
model.summary()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train=scaler.transform(X_train)
X_train
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
enc= OneHotEncoder()
```

```python
enc.fit(y)
```

```python
y_train_enc=np.array(enc.transform(y_train).toarray(),dtype='float32')
y_train_enc
```

```python
y_test_enc = np.array(enc.transform(y_test).toarray(), dtype='float32')
y_test_enc
```

```python
model.compile(loss="categorical_crossentropy",optimizer=Adam(lr=0.1),metrics=['acc'])
```

```python
model.fit(X_train,y_train_enc,epochs=1001)
```

```python
X_test = scaler.transform(X_test)
X_test
```

```python
predict = model.predict(X_test)
```

```python
predict01 = np.where(predict>0.5,1,0)
predict01
```

```python
predict_arg = np.argmax(predict01,1)
```

```python
y_test_enc_arg = np.argmax(y_test_enc,1)
```

```python
acc = (predict_arg == y_test_enc_arg)
acc
```

```python
np.average(acc)
```
