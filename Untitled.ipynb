# 전처리

```python
import missingno as msno
msno.bar(df=X)
```

```python
X=np.array(X,dtype='float32')
X
```

```python
y = np.array(df['interest'],dtype='float32')
y=y.reshape(-1,1)
y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2, random_state = 3282)
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
enc=OneHotEncoder()
```

```python
enc.fit(y)
```

```python
y_train_enc = enc.transform(y_train).toarray()
y_test_enc = enc.transform(y_test).toarray()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
X_train
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
```

```python
model = Sequential()
```

```python
model.add(Dense(200,input_dim=14, activation='sigmoid'))
```

```python
model.add(Dense(100,activation='sigmoid'))
```

```python
model.add(Dense(3,activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(loss='categorical_crossentropy',optimizer=Adam(lr=0.01), metrics=['acc'])
```

```python
model.fit(X_train,y_train_enc,epochs=100)
```

```python
X_test = scaler.transform(X_test)
```

```python
predict0 = model.predict(X_test)
predict0
```

```python
y_test_enc
```

```python
predict1=np.argmax(predict0,1)
target = np.argmax(y_test_enc,1)
predict1
```

```python
acc = (predict1==target)
acc
```
