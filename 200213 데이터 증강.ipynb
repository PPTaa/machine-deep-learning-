```
import numpy as np
import matplotlib.pyplot as plt
from keras.preprocessing.image import ImageDataGenerator
```

```
im_width = 220
im_height = 220
```

```
image_gen = ImageDataGenerator(rotation_range=180, 
                   width_shift_range=0.2,
                   height_shift_range=0.2,
                   horizontal_flip=True,
                   vertical_flip=True)
```

```
train_dir = './gdrive/My Drive/Colab Notebooks/data/dental_image/train'
```

```
train_gen = image_gen.flow_from_directory(train_dir,
                              target_size=(im_width,im_height),
                              batch_size=1,
                              class_mode='categorical',
                              shuffle=True)
```

```
img,label = train_gen.next()
```

```
img
```

```
img=img/255
```

```
plt.imshow(img[0])
```

```
index = 0
all_generate_img = []

for generate_img in image_gen.flow(img):
    print('='*50)
    print('='*50)
    all_generate_img.append(generate_img)
    index += 1 
    if index > 20:
        break
```

```
all_generate_img
```

```
fig = plt.figure(figsize=(20,10))
for i in range(20):
    subplot = fig.add_subplot(4,5,i+1)
    subplot.set_xticks([])
    subplot.set_yticks([])
    subplot.imshow(all_generate_img[i][0])
plt.show()
```
