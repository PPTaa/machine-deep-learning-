# Gradient Decent를 이용한 회귀분석

```python
import numpy as np
```

```python
X=np.array([1,2,3],dtype='float32')
X
```

```python
y=np.array([2,2.5,3.5],dtype='float32')
y
```

```python
w=2
b=1
```

```python
hypothesis = X*w+b
hypothesis
```

```python
hypothesis - y
```

```python
X
```

# Gradient Decent를 이용한 선형회귀분석

```python
w_gred = 1/3*np.dot((hypothesis-y),X)
w_gred
```

```python
w = w - w_gred
w
```

```python
b_gred = 1/3*np.sum(hypothesis-y)
b_gred
```

```python
b = b - b_gred
b
```

```python
for i in range(100):
    hypothesis = X*w+b
    MSE = np.sum((hypothesis-y)**2)
    print('MSE : ',MSE)
    w_gred = 1/3*np.dot((hypothesis-y),X.T)
    print('w_gred : ', w_gred)
    b_gred = 1/3*np.sum(hypothesis-y)
    print('b_gred : ', b_gred)
    w=w-w_gred
    b=b-b_gred
```

# Learning Rate 와 Gradient Decent

```python
w = 2
b = 1
learning_rate = 0.1
```

```python
for i in range(10000):
    hypothesis = X*w+b
    MSE = np.sum((hypothesis-y)**2)
    print('MSE : ',MSE)
    w_gred = 1/3*np.dot((hypothesis-y),X.T)
    print('w_gred : ', w_gred)
    b_gred = 1/3*np.sum(hypothesis-y)
    print('b_gred : ', b_gred)
    w=w-(learning_rate*w_gred)
    b=b-(learning_rate*b_gred)
    print('w : ',w,'b : ',b)
```
