# 인공 신경망을 이용한 AND연산 (파이썬)

```python
import numpy as np
import tensorflow as tf
tf.enable_eager_execution()
```

```python
X=np.array([
            [0,0],
            [0,1],
            [1,0],
            [1,1]
], dtype='float32')
X
```

```python
y=np.array([
            [0],
            [0],
            [0],
            [1]
], dtype='float32')
y
```

```python
w = np.array([[1],[2]], dtype='float32')
w
```

```python
b = np.array([3], dtype='float32')
b
```

```python
np.dot(X,w)
```

```python
np.dot(X,w)+b
```

```python
def sigmoid(h):
    return 1/(1+np.exp(-h))
```

```python
hypothesis = np.dot(X,w)+b
hypothesis
```

```python
hypothesis = sigmoid(np.dot(X,w)+b)
hypothesis
```

```python
hypothesis-y
```

```python
X.T
```

```python
X.T[0]
```

```python
X1=X.T[0].reshape(1,-1)
print(X1)
X2=X.T[1].reshape(1,4)
print(X2)
```

```python
np.dot(X1,(hypothesis-y))
```

```python
np.sum(np.dot(X1,(hypothesis-y)))
```

```python
w1_gred=np.sum(np.dot(X1,(hypothesis-y)))/4
w1_gred
```

```python
w2_gred=np.sum(np.dot(X2,(hypothesis-y)))/4
w2_gred
```

```python
hypothesis-y
```

```python
b_gred=np.sum(hypothesis-y)/4
b_gred
```

```python
learning_rate = 0.1
```

```python
for i in range(100):
    for j in range(2):
        hypothesis=sigmoid(np.dot(X,w)+b)
        print('hypothesis : ',hypothesis)
        cost=1/4*(-y*np.log(hypothesis))-((1-y)*np.log(1-hypothesis))
        print('cost : ',cost)
        X_col_2d=X.T[j].reshape(1,-1)
        w_gred = 1/4*np.sum(np.dot(X_col_2d,(hypothesis-y)))
        print(f'w_{i}gred : {w_gred}')
        w[j]=w[j]-learning_rate*w_gred
        print('w : ',w)
        b_gred=1/4*np.sum(hypothesis-y)
        print('b_gred : ',b_gred)
        b = b-learning_rate*b_gred
        print('b : ',b)
```

```python
predict = sigmoid(np.dot(X,w)+b)
predict
```

```python
predict01 = tf.cast(predict > 0.5, dtype=tf.float32)
print('-'*30)
print(predict01)
```

```python
ac01=tf.equal(predict01,y)
ac01
```
