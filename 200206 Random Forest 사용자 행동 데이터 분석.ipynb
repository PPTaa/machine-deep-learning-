```python
feature_name = feature_name_df.iloc[:,1].values.tolist()
feature_name
```

```python
X_train = pd.read_csv('./data/human_activity_X_train.txt',sep='\s+',header=None)
X_test = pd.read_csv('./data/human_activity_X_test.txt',sep='\s+',header=None)
y_train = pd.read_csv('./data/human_activity_y_train.txt',sep='\s+',header=None,names=['action'])
y_test = pd.read_csv('./data/human_activity_y_test.txt',sep='\s+',header=None,names=['action'])
```

```python
X_train.columns = feature_name
X_test.columns = feature_name
```

```python
import numpy as np
import warnings
from sklearn.metrics import accuracy_score, classification_report
from sklearn.ensemble import RandomForestClassifier
```

```python
rf_clf = RandomForestClassifier()
```

```python
rf_clf.fit(X_train,y_train)
```

```python
pred = rf_clf.predict(X_test)
pred
```

```python
accuracy_score(pred,y_test)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
params = {
    'n_estimators':[100],
    'max_depth':[6,8,10,12],
    'min_samples_leaf':[8,12,18],
    'min_samples_split':[8,16,20]
}
```

```python
rf_clf = RandomForestClassifier(n_jobs=10)
rf_clf
```

```python
grid_cv = GridSearchCV(rf_clf, param_grid=params, n_jobs=10)
grid_cv
```

```python
warnings.filterwarnings('ignore')
```

```python
grid_cv.fit(X_train, y_train)
```

```python
print(grid_cv.best_params_)
print(grid_cv.best_score_)
```

```python
rf_clf1 = RandomForestClassifier(n_estimators=300,max_depth=10,min_samples_leaf=8,min_samples_split=8)
```

```python
rf_clf1.fit(X_train,y_train)
```

```python
pred = rf_clf1.predict(X_test)
```

```python
accuracy_score(pred,y_test)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib
```

```python
ftr_importance_values = rf_clf1.feature_importances_
ftr_importance_values
```

```python
ftr_importance = pd.Series(ftr_importance_values,index=X_train.columns)
ftr_importance
```

```python
ftr_top20 = ftr_importance.sort_values(ascending=False)[:20]
ftr_top20
```

```python
plt.figure(figsize=(8,8))
plt.title('top20')
sns.barplot(x=ftr_top20, y= ftr_top20.index)
```
