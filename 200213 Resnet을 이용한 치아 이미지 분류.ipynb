```
im_width = 220
im_height = 220
```

```
train_dir = './gdrive/My Drive/Colab Notebooks/data/dental_image/train'
test_dir = './gdrive/My Drive/Colab Notebooks/data/dental_image/test'
```

```
train_datagen = ImageDataGenerator( rescale=1./255 ,rotation_range=180,width_shift_range=0.2,height_shift_range=0.2,horizontal_flip=True,vertical_flip=True)
```

```
train_generator = train_datagen.flow_from_directory(train_dir,target_size=(im_width,im_height),batch_size=8,class_mode='categorical',shuffle=True)
```

```
test_datagen = ImageDataGenerator(rescale=1./255)
```

```
test_generator = test_datagen.flow_from_directory(test_dir,target_size=(im_width,im_height),batch_size=60,class_mode='categorical',shuffle=False)
```

```
from keras.models import Sequential
from keras.layers import Dense, Dropout,MaxPooling2D,Flatten,Conv2D
from keras.optimizers import Adam
```

```
#강의자료 204p 추가
from keras.applications.resnet50 import ResNet50
conv_layers=ResNet50(weights='imagenet', include_top=False, input_shape=(im_width,im_height,3))
```

```
conv_layers.summary()
```

```
model = Sequential()
```

```
#강의자료 206페이지 추가
model.add(conv_layers)
```

```
#model.add(Conv2D(32,kernel_size=(3,3), input_shape=(220,220,3), activation='relu', padding='SAME'))
```

```
#model.add(Conv2D(64,kernel_size=(3,3), activation='relu', padding='SAME'))
```

```
#model.add(MaxPooling2D(pool_size=2,padding='SAME'))
```

```
model.add(Flatten())
```

```
model.add(Dense(512,activation='relu'))
```

```
model.add(Dense(3, activation='softmax'))
```

```
model.summary()
```

```
model.compile(loss='categorical_crossentropy',optimizer=Adam(lr=0.00001),metrics=['acc'])
```

```
model.fit(train_generator,epochs=50)
```
