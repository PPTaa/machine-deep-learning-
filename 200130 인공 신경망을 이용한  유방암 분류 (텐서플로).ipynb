```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
import numpy as np
import tensorflow as tf
```

```python
tf.enable_eager_execution()
```

```python
data=datasets.load_breast_cancer()
data
```

```python
X = data.data
X.shape
```

```python
X=np.array(X,dtype='float32')
```

```python
X
```

```python
y=data.target
y=np.array(y,dtype='float32')
print(y.shape)
y
```

```python
y=y.reshape(569,1) # 1차원 배열을 2차원 배열로 변환
print(y.shape)
y
```

```python
X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.2,random_state=1) 
#X_train에 X중80%, X_test에 X중20%, y_train에 y중80%, y_test에 y중20% 로 나누어지는것
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
X_train
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaler.fit(X_train)
```

```python
X_train=scaler.transform(X_train)
X_train
```

```python
initializer = tf.contrib.layers.xavier_initializer()
w=tf.Variable(initializer([30,1]))
print(w)
b=tf.Variable(initializer([1]))
print(b)
```

```python
tf.matmul(X_train,w)+b
```

```python
arr = X_train[0]
arr=arr.reshape(1,30)
arr
```

```python
tf.matmul(arr,w)+b
```

```python
hypothesis=tf.sigmoid(tf.matmul(X_train,w)+b)
hypothesis
```

```python
optimizer = tf.train.AdamOptimizer(0.001)
```

```python
for i in range(600):
    with tf.GradientTape() as tape:
        hypothesis = tf.sigmoid(tf.matmul(X_train,w)+b)
        cost = -tf.reduce_mean(y_train*tf.log(hypothesis)+(1-y_train)*tf.log(1-hypothesis))
        print(f'cost{i} :',cost)
        greds = tape.gradient(cost,[w,b])
        print(greds)
    optimizer.apply_gradients(grads_and_vars=zip(greds,[w,b]))
    print('---'*30)
```

```python
X_test=scaler.transform(X_test)
X_test
```

```python
predict = tf.sigmoid(tf.matmul(X_test,w)+b)
predict
```

```python
predict01 = tf.cast(predict>0.5,dtype='float32')
predict01
```

```python
tf.reduce_mean(tf.cast(tf.equal(predict01,y_test),dtype='float32'))
```
