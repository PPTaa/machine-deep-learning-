```python
lemm_dic = {
    'NN':'n',
    "JJ":'a',
    'VB':'v',
    'RB':'r'
}
lemm_dic
```

```python
def clearText(description):
    tokens = nltk.word_tokenize(description)
    tokens = [token.lower() for token in tokens if len(token)>1]
    tokens_clean = [token for token in tokens if not token in stop_words]
    tokens_tagged = nltk.pos_tag(tokens_clean)
    tokens_tagged = [(word, pos[0:2]) 
                     for word, pos in tokens_tagged 
                     if pos[0:2] in ['NN','JJ','VB','RB']]
    tokens_tagged = [lem.lemmatize(word, lemm_dic[pos]) for word, pos in tokens_tagged]
    return " ".join(tokens_tagged)
```

```python
df['description'][0]
```

```python
clearText(df['description'][0])
```

```python
df['description_clear'] = df['description'].apply(clearText)
df['description_clear']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df['description_clear'],df['category'], test_size=0.2, random_state=156)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np
```

```python
vectorizer = TfidfVectorizer(min_df=2, ngram_range=(1,2), 
                             strip_accents='unicode',norm='l2')
```

```python
X_train_vector = vectorizer.fit_transform(X_train)
X_train_vector
```

```python
X_train_df = pd.DataFrame(X_train_vector.toarray(), columns=vectorizer.get_feature_names())
X_train_df
```

```python
X_test_vector = vectorizer.transform(X_test)
```

```python
X_test_df = pd.DataFrame(X_test_vector.toarray(), columns=vectorizer.get_feature_names())
X_test_df
```

```python
from xgboost import XGBClassifier
```

```python
xgb_clf = XGBClassifier(n_estimators=30)
xgb_clf
```

```python
params = {'max_depth':[3,5,7],
         'min_child_weight':[3,5,7],
         'colsample_bytree':[0.1,0.5,1.0],
         'learning_rate':[0.001,0.01,0.1],
         'num_class':[8]}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
gridcv = GridSearchCV(xgb_clf, param_grid=params)
gridcv
```

```python
gridcv.fit(X_train_vector, y_train, eval_set=[(X_test_vector, y_test)])
```

```python
gridcv.best_params_
```

```python
pred = gridcv.predict(X_test_vector)
pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
print(accuracy_score(y_test, pred))
```

```python
from sklearn.metrics import confusion_matrix
```

```python
print(confusion_matrix(y_test, pred))
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, pred))
```
